```python
import folium
```

```python
# list for favorite locations and zoom level
# format: list [long, lat], int 1 - 10

scandinavia = [[62.32,17.43], 5]
denmark = [[55.84,10.60], 7]
christiania = [[55.676862, 12.595539], 10]
unitedstates = [[40,-99], 4]
```

```python
import pandas
df = pandas.read_csv('volcanoes.csv')
```

```python
lat = list(df['LAT']) # latitudes from dataframe
lon = list(df['LON']) # longitudes from dataframe
vol_names = list(df['NAME']) # names of volcanoes from dataframe
elev = list(df['ELEV']) # elevation of volcanos from dataframe
```

```python
def marker_color(elevation):
    if int(elevation) < 1500:
        return 'green'
    elif int(elevation) < 2700:
        return 'orange'
    else:
        return 'red'
```

```python
map = folium.Map(location=unitedstates[0], zoom_start=unitedstates[1], tiles='Mapbox Bright')
```

```python
fg = folium.FeatureGroup(name='My map')
```

```python
for lt, ln, vol, el in zip(lat, lon, vol_names, elev):
    fg.add_child(folium.CircleMarker(location=[lt,ln], radius=6, popup=vol + '\n' + str(el) + ' m.', fill_color=marker_color(el), color='grey', weight=1, fill=True, fill_opacity=0.7))
```

```python
# adding polygon map with overlay of countries
fg.add_child(folium.GeoJson(data=open('world.json', 'r', encoding='utf-8-sig').read(),
                                       style_function=lambda x: {
                                           'fillColor': 'green' if x['properties']['POP2005'] < 10000000
                                                   else 'yellow' if 10000000 <= x['properties']['POP2005'] < 20000000
                                                   else 'orange' if 20000000 <= x['properties']['POP2005'] < 50000000
                                                   else 'red'
                                       }))
# lambda function colors polygon countries based on the population
```

```python
map.add_child(fg)
```

```python
map.add_child(folium.LayerControl()) # Layer control must be added after feature group
```

```python
map.save('map_volcanos.htm')
```
