```python
import pandas
```

```python
# read directly from URL
df6 = pandas.read_csv('http://pythonhow.com/supermarkets.csv')
df6.set_index('ID')
```

```python
df6['geo_address'] = df6['Address'] + ', ' + df6['City'] + ', ' + df6['State'] + ', ' + df6['Country']
df6
```

```python
from geopy.geocoders import ArcGIS
```

```python
nom = ArcGIS()
```

```python
df6['Coordinate'] = df6['Address'].apply(nom.geocode)
df6
```

```python
df6['Latitude'] = df6['Coordinate'].apply(lambda x: x.latitude if x != None else None)
df6['Longitude'] = df6['Coordinate'].apply(lambda y: y.longitude if y != None else None)
df6
```
