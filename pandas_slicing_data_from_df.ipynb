```python
# read directly from URL
df6 = pandas.read_csv('http://pythonhow.com/supermarkets.csv')
df6.set_index('ID')
```

```python
# label based accessing of data using df.loc method
df6.loc["2":"3","Address":"Country"] 
```

```python
# index based accessing of data using df.iloc method
df6.iloc[1:3,1:3] # note that with label based indexing the last value is included - not with index based
```

```python
# accessing data with a combination of labels and indexes using df.ix method
df6.ix[3,"Adress":"Employees"]
```
